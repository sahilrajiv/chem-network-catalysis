```python
import pandas as pd
import numpy as np
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from mordred import Calculator, descriptors
from IPython.display import display_html
from itertools import chain,cycle
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
def load_data(reaction_name):
    '''loads the files for statistical analysis of the graph network'''
    old=pd.read_csv(f'network_stats/{reaction_name}_old.csv', delimiter=',', ) 
    new=pd.read_csv(f'network_stats/{reaction_name}_new.csv', delimiter=',',)
    return [old, new]


def sort_data(list_of_dfs, column):
    '''sorts the datasets using a statistical measure'''
    list_of_dfs[0]=list_of_dfs[0].sort_values(by=[column], ascending=False)
    list_of_dfs[1]=list_of_dfs[1].sort_values(by=[column], ascending=False)
    return list_of_dfs

def print_mols(number,list_of_dfs):
    '''prints the top molecules'''

    molecules_old=[]
    for x in list_of_dfs[0].head(number)['Id']:
        molecules_old.append(Chem.MolFromSmiles(x))

    molecules_new=[]
    for x in list_of_dfs[1].head(number)['Id']:
        molecules_new.append(Chem.MolFromSmiles(x))
        
    return [molecules_old,molecules_new]


def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h2>{title}</h2>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td></th>'
    display_html(html_str,raw=True)
    
    
def image_add(df, fname):
    for index, row in df.iterrows():
    #     print(index)
            m=Chem.MolFromSmiles(row['Id'])
            filename=fname+"/"+row['Id']+".png"
            df.loc[index,'image']=filename
            Draw.MolToImage(m).save(filename)
    return df

def plotter(data, prop):
    data1=sort_data(data,prop)
    
    old=data[0][['Id', prop]].head(10)

    new=data[1][['Id', prop]].head(10)
    print(prop)
    plt.figure(figsize=(10,8))
    ax=sns.barplot(x='Id', y=prop, data=old)
    for i in ax.containers:
        ax.bar_label(i,)
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    ax.set_title('Original Network')
    plt.figure(figsize=(10,8))
    ax1=sns.barplot(x='Id', y=prop, data=new)
    for i in ax1.containers:
        ax1.bar_label(i,)
    ax1.set_xticklabels(new.Id,rotation = 30)
    ax1.set_title('Iron Inclusive Network')

    

    plt.show()
    
def cleaner(data_list):
    data_list[1]=data_list[1].drop(['Label','timeset'], axis=1)
    data_list[0]=data_list[0].drop(['Label','timeset'], axis=1)


    data_list[0]=data_list[0].loc[data_list[0].type==1,:].reset_index(drop=True)
    data_list[1]=data_list[1].loc[data_list[1].type==1,:].reset_index(drop=True)
    
    return data_list
```

# Runtime

```python
formose_data=cleaner(load_data('formose_1608'))
glucose_data=cleaner(load_data('glucose'))
maillard_data=cleaner(load_data('maillard'))
pyruvic_data=cleaner(load_data('pyruvic'))
```

```python
def percentage_calc(dataset):
    for x in dataset:
        for col in ['weighted indegree','weighted outdegree','Weighted Degree']:
            x[f'{col}_%']=x[col]*100/x[col].sum()
    common=pd.merge(dataset[0],dataset[1],how='inner', on='Id')
    return common
py_new=percentage_calc(pyruvic_data)
f_new=percentage_calc(formose_data)
g_new=percentage_calc(glucose_data)
m_new=percentage_calc(maillard_data)


```
